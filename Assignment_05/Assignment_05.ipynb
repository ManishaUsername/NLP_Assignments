4: Define a Function to Get Synonyms

```python
def get_synonyms(word):
    synonyms = set()
    for synset in wn.synsets(word):
        for lemma in synset.lemmas():
            synonyms.add(lemma.name())
    return synonyms
```

5: Define a Function to Get Antonyms

```python
def get_antonyms(word):
    antonyms = set()
    for synset in wn.synsets(word):
        for lemma in synset.lemmas():
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())
    return antonyms
```

6: Define a Function to Get Hypernyms

```python
def get_hypernyms(word):
    hypernyms = set()
    for synset in wn.synsets(word):
        for hypernym in synset.hypernyms():
            for lemma in hypernym.lemmas():
                hypernyms.add(lemma.name())
    return hypernyms
```

7: Apply Functions to a Sample Word

```python
word = "dog"

print("Word:", word)
print("Synonyms:", get_synonyms(word))
print("Antonyms:", get_antonyms(word))
print("Hypernyms:", get_hypernyms(word))
```

8: Apply to Text Data (Multiple Words)

```python
text = "The dog is a faithful animal"
words = text.lower().split()

results = {}
for word in words:
    results[word] = {
        "synonyms": get_synonyms(word),
        "antonyms": get_antonyms(word),
        "hypernyms": get_hypernyms(word)
    }

results
```

9: Pretty Print Results

```python
for word, relations in results.items():
    print(f"\nWord: {word}")
    print("  Synonyms:", list(relations["synonyms"])[:5])
    print("  Antonyms:", list(relations["antonyms"]))
    print("  Hypernyms:", list(relations["hypernyms"])[:5])
```
